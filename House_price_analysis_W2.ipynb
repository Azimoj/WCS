**Filing missing values**

```python
## Now lets check for numerical variables the contains missing values
numerical_with_nan=[feature for feature in train_clean.columns if train_clean[feature].isnull().sum()>1 and train_clean[feature].dtypes!='O']

## We will print the numerical nan variables and percentage of missing values

for feature in numerical_with_nan:
    print("{}: {}% missing value".format(feature,np.around(train_clean[feature].isnull().mean(),4)))
```

```python
## Replacing the numerical Missing Values

for feature in numerical_with_nan:
    ## We will replace by using median since there are outliers
    median_value=train_clean[feature].median()

    ## create a new feature to capture nan values
    train_clean[feature+'nan']=np.where(train_clean[feature].isnull(),1,0)
    train_clean[feature].fillna(median_value,inplace=True)

train_clean[numerical_with_nan].isnull().sum()
```

```python
categorical_features=[feature for feature in train_clean.columns if train_clean[feature].dtype=='O']
```

```python
for feature in categorical_features:
    temp=train_clean.groupby(feature)['SalePrice'].count()/len(train_clean)
    temp_df=temp[temp>0.01].index
    train_clean[feature]=np.where(train_clean[feature].isin(temp_df),train_clean[feature],'Rare_var')
```

**Converting cathegorical values to ordinal values**

```python
train_lable_order=train_clean.copy()
```

```python
for feature in categorical_features:
    labels_ordered=train_lable_order.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    train_lable_order[feature]=train_lable_order[feature].map(labels_ordered)
```

```python
import pandas as pd
from statsmodels.stats import weightstats as stests

ztest, pval1 = stests.ztest(x1=train[train['MSZoning']=='RM']['SalePrice'], x2=train[train['MSZoning']=='C (all)']['SalePrice'],  value=0, alternative='larger')

print(float(pval1))

if pval1 < 0.05:
    print("reject null hypothesis,le prix moyen des maisons en zone à densité moyenne est significativement supérieur au prix des maisons dans les zones commerciales.")
else:
    print("accept null hypothesis")
```

Ha: Différence significative





```python
import scipy.stats as stats
from scipy.stats import f_oneway

# On va faire une one-way ANOVA

df_anova = train[['MSZoning','SalePrice']]

grps = pd.unique(df_anova.MSZoning.values)
d_data = {grp:df_anova['SalePrice'][df_anova.MSZoning == grp] for grp in grps}

F, p = stats.f_oneway(d_data['RL'], d_data['RM'], d_data['C (all)'],d_data['FV'],d_data['RH'])

print("p-value for significance is: ", p)

if p<0.05:
    print("reject null hypothesis, le prix des maisons est significativement différent entre les zones résidentielles")
else:
    print("accept null hypothesis")
```
